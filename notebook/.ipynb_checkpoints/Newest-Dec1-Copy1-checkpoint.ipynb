```python
def clean(text, rating):
    """
    1. Convert . to space
    2. Remove all special chacaters.
    3. Remover more than one space
    """
    text = re.sub('\.', ' ', text)
    text = re.sub('[^A-Za-z ]+', '', text)
    text = re.sub(' +', ' ', text)
    
    '''
    Rating feature is included into the text
    Not added 
    '''
    
#     if rating == 5.0:
#         text += ' 5rating'
#     elif rating >= 4.0:
#         text += ' 4rating'
#     elif rating >= 3.0:
#         text += ' 3rating'
#     elif rating >= 2.0:
#         text += ' 2rating'
#     else:
#         text += ' 0rating'
    
    return text
        
```

```python
def process_ds():
    '''
    B is budget
    '''
    
    X = []
    y = []
    for i in range(len(data['reviews'])):
        if data['reviews'][i]['overall'] >= 5: 
            X.append(clean(data['reviews'][i]['reviewText'],data['reviews'][i]['overall']))
            y.append(1)
        else:
            X.append(clean(data['reviews'][i]['reviewText'], data['reviews'][i]['overall'])) 
            y.append(0)
    print(f"Shape : X = {pd.DataFrame(X).shape}, y = {pd.DataFrame(y).shape}")
    return pd.DataFrame(X), pd.DataFrame(y)
```

```python
def load_dataset(B , X, y):
    '''
    B is budget
    '''
    
    vectorizer = CountVectorizer(min_df=5, max_df=1.0, binary=True)
    new_X = vectorizer.fit_transform(X.iloc[:B, 0])

    new_X = new_X.todense()
    XX_train, XX_test, yy_train, yy_test = train_test_split(new_X, y, test_size=0.33, random_state = 44)
    print(XX_train, XX_test, yy_train, yy_test)
    return ""
```

```python
def get_theta(model):
    '''
    returns trained theta
    '''
    stats = {}
    if model == 'BNB':
        clf = BernoulliNB()
    if model == 'LR':
        clf = LogisticRegression(penalty='l2')
    if model == 'DT':
        clf = tree.DecisionTreeClassifier()
    
    clf.fit(X_train, y_train)
    
    #clf = BernoulliNB()
    #clf.partial_fit(X_train[0:40], y_train[0:40], [0, 1])
    #clf.partial_fit(X_train[40:], y_train[40:],  [0, 1])
    
    y_predict = clf.predict(X_test)

    y_pred_proba = clf.predict_proba(X_test)[::,1]
    stats['roc_curve'] = metrics.roc_curve(y_test,  y_pred_proba)
    stats['auc'] = metrics.roc_auc_score(y_test, y_pred_proba)
    stats['accuracy'] = accuracy_score(y_test, y_predict)
    stats['clf'] = clf
    stats['B'] = X.shape[0]
    print(f"Completed training {model}")
    return stats
```

```python
'''
step 1: It converts the co-ef to absolute value
step 2: Appends to info list(feature_name, weight)
'''
def top_weighted_features(coef,label):
    coef = np.ravel(np.abs(coef))
    info = []
    index = np.ravel(coef).argsort()[::-1][:11]
    for t in range(11):
        info.append([X.columns[index[t]],coef[index[t]]])
    return pd.DataFrame(info,columns=['Features('+label+')','Weights('+label+')'])
```

```python
def train_several(model):
    b = 0
    t = 40
    thetas = []
    for i in range(t):
        b += 50
        X, y, f = load_dataset(b)
        thetas.append(get_theta(model, X, y))
    return thetas
```
